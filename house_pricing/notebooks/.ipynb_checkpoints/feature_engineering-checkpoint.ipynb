```python
# Plot price per sqm vs. the price. Some clear outliers here: low priced properties of suspiciously small size.
fig, ax = plt.subplots(1,2,figsize=(7,4))
ax[0].scatter(Train['price_doc'], Train['price_doc'] / Train['full_sq'], color='teal', alpha=0.4)
ax[0].set_xlabel('price')
ax[0].set_ylabel('price per sqm')
ax[1].scatter(Train['price_doc'], Train['price_doc'] / Train['full_sq'], color='teal', alpha=0.4)
ax[1].set_ylim([0, 1e6])
ax[1].set_xlabel('price')
```

```python
# There is a nice linear relationship between the number of rooms and full_sq for room numbers in the range of [1, 9]:
fig, ax = plt.subplots(1,1,figsize=(3,3))
ax.scatter(np.sort(Train['num_room'].unique()[1:]), Train.groupby('num_room')['full_sq'].mean(), color='teal', alpha=0.4)
ax.set_xlabel('num_room')
ax.set_ylabel('average_full_sq')
```
