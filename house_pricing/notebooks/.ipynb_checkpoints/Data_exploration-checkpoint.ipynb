# Correlation analysis

```python
threshold = 0.8

macro_correlation_matrix = macro.drop('timestamp', axis=1).corr().abs()
train_correlation_matrix = Train.drop('timestamp', axis=1).corr().abs()

macro_tria = macro_correlation_matrix.where(np.triu(np.ones(macro_correlation_matrix.shape), k=1).astype(bool))
train_tria = train_correlation_matrix.where(np.triu(np.ones(train_correlation_matrix.shape), k=1).astype(bool))

#Features to drop from Train and macro (optional, of course) based on a correlation threshold:
to_drop_macro = [column for column in macro_tria.columns if any(macro_tria[column] > threshold)]
to_drop_train = [column for column in train_tria.columns if any(train_tria[column] > threshold)]
```

```python
with open(path_2 + "to_drop_macro", "wb") as fp:
    pickle.dump(to_drop_macro, fp)
    
with open(path_2 + "to_drop_train", "wb") as fp:
    pickle.dump(to_drop_train, fp)
```

```python
# Look at the 5 most correlated features with target.
top_5_corr = train_correlation_matrix['price_doc'].sort_values().iloc[-6:-1].index
fig, ax = plt.subplots(1,len(top_5_corr),figsize=(16,3))

i = 0
for feature in top_5_corr:
    ax[i].scatter(Train[feature], Train['price_doc'], color='teal')
    ax[i].set_xlabel(feature)
    ax[0].set_ylabel('Price')
    i += 1
```

Clear otulier in 'full_sq'- the point with ~5000. Also, a few suspicious points in 'num_room': properties with more than 10 rooms are in the cheapest category.

# Feature importance from xgboost

Here I'll check the feature importance by fitting an xgboost model.

```python
# First need to encode the categoricals. For now lets use ordinal encoding.
for feature in categorical_features:
    combined_train[feature] = pd.factorize(combined_train[feature])[0]
    
# Replace timestamp with year:
combined_train['timestamp'] = pd.DatetimeIndex(combined_train['timestamp']).year
```

```python
# Train xgboost model using entire set:

model = XGBRegressor(eta=0.03, max_depth=5, subsample=1.0, colsample_bytree=0.7, objective='reg:squarederror',
                    eval_metric='rmse', silent=1)

model.fit(combined_train, np.log1p(target_train))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
xgb.plot_importance(model, max_num_features=20, height=0.5, ax=ax, color='teal')
```

```python
# Sorted dictionary (lowest to largest) of importance values:
important_features = model.get_booster().get_score(importance_type='weight')
important_features_sorted = {k: v for k, v in sorted(important_features.items(), key=lambda item: item[1])}

top8 = list(important_features_sorted.keys())[-8:]
```

```python
top100 = list(important_features_sorted.keys())[-100:]

with open(path_2 + "top_100", "wb") as fp:
    pickle.dump(top100, fp)
```

```python
# How many of the 30 most important features are from the macro data set:
nf = sum([1 for f in list(important_features_sorted.keys())[-30:] if f in macro.columns.tolist()])
print(nf)
```

```python
# Scatter matrix for the 8 most imporant features:
sm = scatter_matrix((combined_train[top8].
                     query('full_sq < 1000').
                     query('life_sq < 1000').
                     query('build_year < 2020 or build_year > 1500')), 
                    color='maroon', figsize=(20,20), marker='o', diagonal='hist')
```

```python
# Train xgboost model without the 'full_sq' feature, which otherwise gets an overwhelming amount of importance:

model = XGBRegressor(eta=0.03, max_depth=5, subsample=1.0, colsample_bytree=0.7, objective='reg:squarederror',
                    eval_metric='rmse', silent=1)

model.fit(combined_train.drop('full_sq', axis=1), np.log1p(target_train))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
xgb.plot_importance(model, max_num_features=20, height=0.5, ax=ax, color='purple')
```
